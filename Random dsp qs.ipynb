(1) Suppose you sample an analog signal $x(t)=1-t$ during the interval $0\leqslant t \leqslant 1 sec$ with a
sampling period of $0.125$ to generate $x[n]$. Plot $x[n]$, $x[-n]$, $x_{even}[n]$ and $x_{odd}[n]$.

```python
Fs = 0.125
n = np.arange(0,1.125,Fs)
x_n = 1 - n
plt.subplot(2,2,1)
plt.stem(n,x_n)
plt.title('x[n]')

x_neg = np.flip(x_n)
n_neg = np.arange(-1,0.125,Fs)
plt.subplot(2,2,3)
plt.stem(n_neg, x_neg)
plt.title('x[-n]')

z = np.zeros(8)
x_n = np.append(z,x_n)
x_neg = np.append(x_neg,z)

x_even = ( x_n + x_neg)/2
x_odd = (x_n - x_neg)/2

N = np.arange(-1,1.125,Fs)
plt.subplot(2,2,2)
plt.stem(N, x_even)
plt.title('x[n]even')

plt.subplot(2,2,4)
plt.stem(N,x_odd)
plt.title('x[n]odd')

plt.show()
```

(2)Let $x[n]=[2, 4, -3, \textbf{1}, -5, 4, 7]$ $-3\leqslant n \leqslant 3$. Generate and plot the samples of the 
following sequences using stem function.

\begin{equation}\nonumber
(i)~~~~x_{1}[n] = 2x[n-3] + 3x[n + 4] - x[n]
\end{equation}
\begin{equation}\nonumber
(ii)~~~~~~~~~~~x_{2}[n] = x[-n- 4] + x[-n + 4]
\end{equation}
\begin{equation}\nonumber
(iii)~~~~~~~~~~~~~~x_{3}[n] = 2e^{0.5n} x[n] u[-n-1] 
\end{equation}

```python
x_n = [0,0,0,0,2,4,-3,1,-5,4,7,0,0,0]
n = np.arange(-7,7)
plt.rcParams["figure.figsize"] = (12, 8)
plt.subplot(2,2,1)
plt.stem(n,x_n)
plt.title('x[n]')
a = 2*np.roll(x_n,3) + 3*np.roll(x_n,-4) - np.flip(x_n)
plt.subplot(2,2,2)
plt.stem(n,a)
plt.title('x1[n]')

x_n = [0,0,0,0,2,4,-3,1,-5,4,7,0,0,0,0]
n = np.arange(-7,8)
sub = np.roll(x_n,4) #x[n-4]
sub = np.flip(sub)
sub2 = np.roll(x_n,-4) #x[n+4]
sub2 = np.flip(sub2)
b = sub + sub2
plt.subplot(2,2,3)
plt.stem(n,b)
plt.title('x2[n]')

x_n = [2,4,-3,1,-5,4,7]
n = np.arange(-3,4)
u = []
for i in n:
    if i<=-1:
        u.append(1)
    else:
        u.append(0)
c = 2*np.exp(0.5*n)*x_n*u
plt.subplot(2,2,4)
plt.stem(n,c)
plt.show()
```

(3) Consider the discrete time signal $x[n]=\cos(\frac{2\pi n}{7})$. \
(i) Obtain and plot the downsampled (decimated) signal $y[n]=x[2n]$ along with $x[n]$\
(ii) Obtain and plot the upsampled (interpolated) signal $z[n]=x[n/2]$ for $n$ even and $z[n]=0$ otherwise along with $x[n]$

```python
n = np.arange(-10,11)
arg = 2*np.pi*n/7
x_n = np.cos(arg)

plt.rcParams["figure.figsize"] = (12, 8)
plt.subplot(2,1,1)
plt.stem(n,x_n,'--b',markerfmt='bo',label = 'x[n]')


n = np.arange(-5,6)
arg = 2*np.pi*n/7
y_n = np.cos(arg)

plt.stem(n,y_n,'g', markerfmt='go', label = 'y[n]')
plt.legend()
plt.show()

print('(ii)')
n = np.arange(-20,21)
arg = 2*np.pi*n/7
k_n = np.cos(arg)
z_n = []
for i in n:
    if i%2==0:
        z_n.append(k_n[i])
    else:
        z_n.append(0)
p = np.arange(-10,11)
plt.subplot(2,1,2)
plt.stem(p,x_n,'--b',markerfmt='bo',label = 'x[n]')
plt.stem(n,z_n,'g', markerfmt='go', label = 'z[n]')
plt.legend()
plt.show()
```
