Pyplot


```
import matplotlib.pyplot as plt
import numpy as np
```

```
# plot() takes 2 parameters - points on x-axis, points on y-axis 
xpoints = np.array([0,6])
ypoints = np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```
# plot multiple points
xpoints = np.array([1,3,4,8])
ypoints = np.array([6,3,1,8])
plt.plot(xpoints,ypoints)
plt.show()

#plot as points 
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```
# plot using markers 
plt.plot(xpoints,ypoints,marker ='D')
plt.show()
```

```
# format strings 
# marker|line|color
plt.plot(xpoints,ypoints,'o:r')
plt.show()
```

```
plt.plot(xpoints,ypoints,color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.show()
```

```
# marker size and color
plt.plot(xpoints,ypoints,marker ='o',ms=20, mec='r')
plt.show()
# mec and mfc - arguments for marker color
plt.plot(xpoints,ypoints,marker ='D',ms=20,mfc='r')
plt.show()
```

```
# multiple lines
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])
plt.plot(y1) # x -default
plt.plot(y2)
plt.show()
```

```
# Labels 
xpoints = np.array([1,3,4,8])
ypoints = np.array([6,3,1,8])
plt.plot(xpoints,ypoints)

#.xlabel(), .ylabel(), .title()
# use fontdict argument for fonts
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

#use loc parameter in title() to position the title

plt.xlabel('x axis',fontdict=font2)
plt.ylabel('y-axis',fontdict=font2)
plt.title('graph title',fontdict=font1,loc='left')
plt.show()
```

```
# Grid
font1 = {'family':'Times New Roman','color':'white','size':20}
font2 = {'family':'Times New Roman','color':'darkred','size':15}

plt.xlabel('x axis',fontdict=font2)
plt.ylabel('y-axis',fontdict=font2)
plt.title('graph title',fontdict=font1)

plt.plot(xpoints,ypoints)
plt.grid()
plt.show()
```

```
plt.plot(xpoints,ypoints)
plt.grid(axis = 'x')
plt.show()
```

```
# line properties to grid
plt.plot(xpoints,ypoints)
plt.grid(color='purple',linestyle='--',linewidth=0.5)
plt.show()
```

Subplots

```
#plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1) # 1 row, 2 coloumns , first plot
plt.plot(x,y)

#plot 2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2) # 1 row, 2 columns , second plot
plt.plot(x,y)

plt.show()
```

```
#plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,1,1) 
plt.plot(x,y)
plt.title('plot-1')

#plot 2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,1,2) 
plt.plot(x,y)
plt.title('plot-2')

plt.suptitle('my plots')
plt.show()
```

Bars and charts

```
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.bar(x,y,width=0.1)
plt.show()
```

```
# horizontal bars
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.barh(x,y,color='hotpink')
plt.show()
```

```
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```
y = np.array([35,25,25,15])
mylabels = ['color1','color2','color3','color4']
plt.pie(y,labels=mylabels)
plt.legend(title='colors')
plt.show()
```
